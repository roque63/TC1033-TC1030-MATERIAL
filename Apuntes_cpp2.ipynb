# **1.1¬∫ Declaraci√≥n del Arreglo**

La sintaxis b√°sica para declarar un arreglo en C++ es:

```cpp
tipo_de_dato  nombreDelArreglo[size];
```
- Donde:
  - `tipo_de_dato`: Es el tipo de datos de los elementos del arreglo (por ejemplo, int, char, double, etc.).
  - `nombreDelArreglo`: Es el nombre que le das al arreglo.
  - `size`: Es la cantidad de elementos que puede contener el arreglo.

Ejemplo: declaraci√≥n de un arreglo de enteros con 35 elementos, se usa la siguiente sintaxis:
```cpp
int calificaciones[35];
```

Ejemplo: declaraci√≥n e inicializaci√≥n de un arreglo de enteros con 10 elementos, se usa la siguiente sintaxis:
```cpp
                 
int calificaciones2[] = {100, 98, 99, 100, 89, 88, 100, 100, 85, 100}
```

## √çndices para accesar un elemento del arreglo
- Para accesar - **obtener el valor , cambiar el valor**- de un elemento del arreglo se utiliza un √≠ndice.
- El √≠ndice tiene que ser de tipo entero.
- Los √≠ndices de un arreglo comienzan desde 0 hasta tama√±o - 1.  Por ejemplo:

```cpp
// Asigna el valor 100 al primer elemento del arreglo - √≠ndice 0
calificaciones[0] = 100;  
```

- Es importante tener en cuenta que los arreglos en C++ son de tama√±o fijo y no pueden cambiar su tama√±o una vez que han sido declarados.

# **1.3 Arreglos - son par√°metros por referencia por default**

### Ejemplo:
https://cplusplus.com/reference/cstdlib/rand/

```cpp
#include <stdlib.h>     /* srand, rand */
#include <time.h>       /* time */

void inicializa(int arrA[], int n){
  // Inicializar la semilla random para que sea diferente la secuencia
  srand (time(NULL));
  for (int index = 0; index < n; index++){
    arrA[index] = rand() % 2000 - 1000;
  }
}

int main(){
  // 1¬∫ Declaraci√≥n del arreglo
  int arrB[10];

  // 2¬∫ llama a la funci√≥n que inicializa los elementos del arreglo
  inicializa(arrB,10);
  
  return 0;
}
```
### Explicaci√≥n:

- En el momento que se lleva a cabo la ejecuci√≥n de la funci√≥n `inicializa`, el arreglo **`arrA`** modifica directamente el **`arrB`**.
- El hecho de que el arreglo **`arrB`** se modifico es porque todos los arreglos tienen un paso de par√°metros por REFERENCIA en forma autom√°tica. ‚úÖ  
- El par√°metro de la funci√≥n apunta a la misma direcci√≥n del argumento correspondiente, y por lo tanto si dentro de la funci√≥n se modifico, estas modificaciones se hacen el argumento usado al llamar a la funci√≥n. üíØ

`````python
%%writefile inicializaRandom.cpp
#include <iostream>
#include <stdlib.h>     /* srand, rand */
#include <time.h>       /* time */
using namespace std;
void inicializa(int arrA[], int n){
  // Inicializar la semilla random para que sea diferente la secuencia
  srand (time(NULL));
  for (int index = 0; index < n; index++){
    arrA[index] = rand() % 2000 - 1000;
  }
}

void despliega(int arrA[], int n){
  // es ineficiente ---
  for (int index = 0; index < n ; index++){
    cout << arrA[index] << ", ";
    if (index == n - 1){
    cout << arrA[index] ;
   }

  for (int index = 0; index < n - 1; index++){
    cout << arrA[index] << ", ";
  }
  if (n > 0){
    cout << arrA[n - 1] << "]\n";
  }
  else{
    cout << "]\n";
  }
}

int main(){
  // 1¬∫ Declaraci√≥n del arreglo
  int arrB[10];

  // 2¬∫ llama a la funci√≥n que inicializa los elementos del arreglo
  inicializa(arrB,10);
  despliega(arrB, 10);
  return 0;
}
`````

# **1.5 Obtener, calcular la cantidad de elementos de un arreglo en C++**  

 - Usa la funci√≥n `sizeof` o calcular el tama√±o del arreglo dividiendo el tama√±o total del arreglo por el tama√±o de un elemento.

 - La funci√≥n `sizeof(arr)` en C++ retorna el tama√±o total del arreglo arr en bytes.   
 Este tama√±o incluye todos los elementos del arreglo y cualquier posible espacio adicional debido a la alineaci√≥n de memoria.

 - La sintaxis general para utilizar sizeof es sizeof(var), donde var es la variable para la cual deseas obtener el tama√±o en bytes.

- Por ejemplo, si tienes un arreglo de enteros int arr[5], y si `sizeof(int)` es 4 bytes en tu plataforma, entonces `sizeof(arr)` te devolver√° el tama√±o total del arreglo en bytes.    
-  Si `sizeof(int)`  es 4 bytes, entonces  `sizeof(arr)` ser√° igual a 5 * 4, es decir, 20 bytes.

- Este valor puede variar dependiendo de la arquitectura y el compilador que est√©s utilizando.



`````python
%%writefile SizeArray.cpp

#include <iostream>
#include <string>
using namespace std;
int main() {
    int size, x;
    int arrA[] = {10, 20, 30, 40, 50};
    string frase = "No dejes para ma√±ana lo que puedas aprender hoy";

    cout << "Ocupaci√≥n en bytes = " << sizeof(arrA) << endl;
    cout << "Ocupaci√≥n en bytes de un elemento del arreglo = " <<  sizeof(arrA[0]) << endl;
    cout << "Elementos del arreglo: " << sizeof(arrA) / sizeof(arrA[0]) << endl;

    size = sizeof(arrA) / sizeof(arrA[0]);
    x = arrA[size - 4];
    cout << x << endl;
    cout << "contenido\n";
    cout << arrA[size] << endl;
    cout << arrA[size -1] << endl;
    cout << arrA[size -2] << endl;
    cout << arrA[size -3] << endl;
    cout << arrA[size -4] << endl;

    return 0;
}
`````

# **2. Matrices o arreglos de 2 dimensiones- Declaraci√≥n**

- Un arreglo de dos dimensiones, tambi√©n conocido como matriz, es una estructura de datos en programaci√≥n que consta de renglones y columnas.
- Es como una tabla en la que cada celda contiene un elemento y est√° identificada por dos √≠ndices: uno para el rengl√≥n y otro para la columna.

En C++, un arreglo de dos dimensiones se declara de la siguiente manera:

```cpp
tipo de dato nombreMatriz[numRenglones][numColumnas];
```

- `tipo de dato`: Es el tipo de datos que contendr√° cada elemento de la matriz (por ejemplo, int, char, double, etc.).
- `nombreMatriz`: Es el nombre que le das al arreglo de dos dimensiones.
- `numRenglones` y `numColumnas`: Son cantidad de renglones y columnas de la matriz.

Por ejemplo, puedes declarar una matriz de enteros 3X3 de la siguiente manera:

```cpp
int matriz[3][3];
```

- Acceso a un elemento espec√≠fico  se deben utilizar dos √≠ndices, uno para la fila o rengl√≥n y otro para la columna:

```cpp
matriz[0][0] = 10;  // Asigna el valor 10 a la posicion 0,0 de la matriz
```

Gr√°ficamente:


| renglon | col_0   | col_1   | col_2 |  
| ------- | -----------  | -----------  | ----------- |  
|0| matriz[0][0] | matriz[0][1] | matriz[0][2] |
|1| matriz[1][0] | matriz[1][1] | matriz[1][2] |
|2| matriz[2][0] | matriz[2][1] | matriz[2][2] |


Un arreglo de dos dimensiones se utiliza para representar estructuras de datos como tablas, im√°genes bidimensionales, mapeos y muchas otras aplicaciones que requieren organizar datos en forma bidimensional.

# String - m√©todos


La clase `std::string` en C++ proporciona una variedad de m√©todos para manipular y trabajar con cadenas de caracteres. A continuaci√≥n, se presentan algunos de los m√©todos m√°s comunes junto con una breve explicaci√≥n de cada uno:

1. **Constructor:**
   - `string()`: Constructor predeterminado que crea una cadena vac√≠a.


2. **Operadores de Asignaci√≥n y Concatenaci√≥n:**
   - `operator=()`: Asigna una cadena a otra.
   - `operator+=()`: Concatena una cadena con otra.

3. **Elemento Acceso:**
   - `operator[]`: Devuelve el car√°cter en la posici√≥n especificada.
  

4. **Iteradores:**
   - `begin()`: Devuelve un iterador al primer car√°cter de la cadena.
   - `end()`: Devuelve un iterador al car√°cter despu√©s del √∫ltimo de la cadena.

5. **Capacidad:**
   - `size() / length()`: Devuelve la longitud de la cadena.
   - `empty()`: Comprueba si la cadena est√° vac√≠a.
  

6. **Modificaci√≥n:**
   - `append()`: Agrega caracteres al final de la cadena.
   - `assign()`: Asigna una nueva cadena a la actual.
   - `insert()`: Inserta caracteres en una posici√≥n espec√≠fica.
   - `erase()`: Elimina caracteres de la cadena.
   - `replace()`: Reemplaza una parte de la cadena con otra cadena.

7. **B√∫squeda y Subcadenas:**
   - `find()`: Encuentra la primera ocurrencia de una subcadena.
   - `rfind()`: Encuentra la √∫ltima ocurrencia de una subcadena.
   - `substr()`: Devuelve una subcadena de la cadena.

8. **Manipulaci√≥n de Caracteres:**
En C++, puedes usar las funciones toupper y tolower de la biblioteca <cctype> para convertir caracteres a may√∫sculas y min√∫sculas, respectivamente.

```cpp

#include <cctype>  //  para toupper y tolower


    string str = "Hola, Mundo!";
````

   - `tolower()`: Convierte los caracteres a min√∫sculas.
   - `toupper()`: Convierte los caracteres a may√∫sculas.

   - `swap()`: Intercambia el contenido de dos cadenas.

La clase `std::string` en C++ proporciona numerosos m√©todos para operar y manipular cadenas de caracteres. Aqu√≠ tienes una lista de algunos de los m√©todos m√°s comunes de la clase `std::string` junto con una breve explicaci√≥n de cada uno:

1. **`length()` y `size()`:** Devuelven el tama√±o (longitud) de la cadena, es decir, el n√∫mero de caracteres que contiene.

```cpp
string str = "Hola";
int len = str.length(); // len es igual a 4
```

2. **`empty()`:** Devuelve `true` si la cadena est√° vac√≠a (no contiene caracteres), y `false` en caso contrario.

```cpp
string str = "Hola";
bool isEmpty = str.empty(); // isEmpty es igual a false
```

3. **`clear()`:** Elimina todos los caracteres de la cadena, dej√°ndola vac√≠a.

```cpp
string str = "Hola";
str.clear(); // Ahora str est√° vac√≠o
```

4. **`append()` y `+=`:** Agregan caracteres o una cadena al final de la cadena actual.

```cpp
string str = "Hola";
str.append(" Mundo"); // str ahora es "Hola Mundo"
str += "!";
```

5. **`insert()`:** Inserta una cadena o caracteres en una posici√≥n espec√≠fica de la cadena actual.

```cpp
string str = "Hola";
```

6. **`erase()`:** Elimina un rango de caracteres de la cadena.

```cpp
string str = "Hola Mundo";

```

7. **`replace()`:** Reemplaza una porci√≥n de la cadena por otra cadena o caracteres.

```cpp

```

8. **`substr()`:** Devuelve una subcadena a partir de una posici√≥n dada y con una longitud determinada.

```cpp
string str = "Hola Mundo";

```

9. **`find()` y `rfind()`:** Buscan la primera (o √∫ltima) aparici√≥n de una subcadena en la cadena y devuelven la posici√≥n donde se encontr√≥. Si no se encuentra, devuelve `std::string::npos`.

```cpp
string str = "Hola Mundo";
```
